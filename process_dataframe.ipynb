```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
```

```python
years = [2020, 2021, 2022, 2023, 2024]
data = pd.read_csv('data/formated data/data_{}.csv'.format(years[0]))
data = data[data['year'] == years[0]]

weather_data = pd.read_csv('data/weather_data/weather_data_{}.csv'.format(years[0]))

for year in years[1:]:
    curr_data = pd.read_csv('data/formated data/data_{}.csv'.format(year))
    curr_data['year'] = year
    curr_data = curr_data[curr_data['year'] == year]
    data = pd.concat([data,curr_data])
    weather_data = pd.concat([weather_data,pd.read_csv('data/weather_data/weather_data_{}.csv'.format(year))])

data = data[['station_id', 'year', 'month', 'day', 'hour', 'ctx-4', 'ctx-3', 'ctx-2', 'ctx-1', 'percentage_docks_available']]
data['date'] = data['year'].astype(int).astype(str)+'/'+data['month'].astype(int).astype(str)+'/' + data['day'].astype(int).astype(str)+'/'+data['hour'].astype(int).astype(str)
data.sort_values(['year', 'month','day','hour'], ascending = True)

weather_data = weather_data[['year','month','day','hour','mm_precip','temperature']]
weather_data['date'] = weather_data['year'].astype(int).astype(str)+'/'+weather_data['month'].astype(int).astype(str)+'/' + weather_data['day'].astype(int).astype(str)+'/'+weather_data['hour'].astype(int).astype(str)
weather_data = weather_data[['mm_precip', 'temperature', 'date']]
```

```python
stations_info = pd.read_csv('data/info bicing/Informacio_Estacions_Bicing.csv')
stations_info = stations_info[['station_id', 'lat','lon', 'altitude', 'post_code', 'capacity']]
stations_info['post_code'] = '0' + stations_info['post_code'].astype(str)
stations_info
```

```python
full_data = data.copy()
```

```python
full_data = pd.merge(full_data, stations_info, left_on = 'station_id', right_on = 'station_id')
full_data.head()
```

```python
full_data = pd.merge(full_data, weather_data, left_on = 'date', right_on = 'date', how = 'left')
full_data
```

```python
full_data['temperature'] = full_data['temperature'].fillna(value = full_data['temperature'].mean())
full_data['mm_precip'] = full_data['mm_precip'].fillna(value = full_data['mm_precip'].mean())
```

```python
codigos_postales_bcn = np.unique(full_data['post_code'])

locations = pd.read_csv('data/locations_data/barna_pics.csv', encoding = 'utf-16').groupby('addresses_zip_code').size().reset_index()
locations.columns = ['post_code', 'rating']
locations['post_code'] = '0'+locations['post_code'].astype(int).astype(str)

locations.head()
```

```python
full_data = pd.merge(full_data, locations, left_on = 'post_code', right_on = 'post_code', how = 'left')
full_data['rating'] = full_data['rating'].fillna(value = full_data['rating'].median())
full_data.head()
```

```python
data_ocio = pd.read_csv('data/locations_data/bcn_ocio.csv', encoding = 'utf-16').groupby('addresses_zip_code').size().reset_index()
data_ocio.columns = ['post_code', 'ocio']
data_ocio['post_code'] = '0'+data_ocio['post_code'].astype(int).astype(str)
data_ocio.head()
```

```python
full_data = pd.merge(full_data, data_ocio, left_on = 'post_code', right_on = 'post_code', how = 'left')
full_data['ocio'] = full_data['ocio'].fillna(value = full_data['ocio'].median())
```

```python
bicis = pd.read_csv('data/locations_data/equipament.csv', encoding = 'utf-16')
carril_bicis = bicis[bicis["name"].str.contains("Carril - Bici")].groupby(['addresses_zip_code']).size().reset_index()
carril_bicis['addresses_zip_code']='0'+carril_bicis['addresses_zip_code'].astype(int).astype(str)
carril_bicis.columns = ['post_code', 'carriles']
carril_bicis.head()
full_data = pd.merge(full_data, carril_bicis, left_on = 'post_code', right_on = 'post_code', how = 'left')
full_data.head()
```

```python
full_data['carriles'] = full_data['carriles'].fillna(value = full_data['carriles'].median())
```

```python
# Corregim el float que hi havia a year
full_data['year'] = full_data['year'].astype(int)
full_data.head()
```

```python
# Convertim la date y/m/d/h a timestamp
full_data['timestamp'] = pd.to_datetime(full_data['date'], format='%Y/%m/%d/%H')

# Inici
start_date = pd.Timestamp('2020-01-01 00:00:00')
full_data['time'] = (full_data['timestamp'] - start_date).dt.total_seconds() // 3600
full_data['time'] = full_data['time'].astype(int)

# Ara fem drop de date i timestamp
full_data.drop(columns=['timestamp', 'date'], inplace=True)
full_data['time'].head()
```

```python
festius = pd.read_csv('festius/Festius_generals_de_Catalunya.csv')
festius
```

```python
festius['Data'] = pd.to_datetime(festius['Data'], format='%d/%m/%Y')

# Extraiem dia, mes i any per poder comparar amb full data
festius['year'] = festius['Data'].dt.year
festius['month'] = festius['Data'].dt.month
festius['day'] = festius['Data'].dt.day
```

```python
full_data = full_data.merge(festius[['year', 'month', 'day']], on=['year', 'month', 'day'], how='left', indicator=True)
full_data['festius'] = (full_data['_merge'] == 'both').astype(int)
full_data.drop(columns=['_merge'], inplace=True)
full_data.head()
```

```python
# Crear una columna 'is_weekend' que valga 1 si es sábado o domingo, 0 si no
full_data['is_weekend'] = pd.to_datetime(full_data[['year', 'month', 'day']]).dt.dayofweek.apply(lambda x: 1 if x >= 5 else 0)
```

```python
years = full_data['year'].unique()

# Guardem full_data filtrat per anys ja que són fitxers grans
for year in years:
    yearly_data = full_data[full_data['year'] == year]
    yearly_data.to_csv(f'../full_data_{year}.csv', index=False)
    print(f'Saved full_data_{year}.csv')
```
