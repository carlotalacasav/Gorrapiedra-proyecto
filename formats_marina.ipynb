```python
data['year'] = data['last_updated'].dt.year
data['month'] = data['last_updated'].dt.month
data['day'] = data['last_updated'].dt.day
data['hour'] = data['last_updated'].dt.hour

data.drop(columns=['last_updated'], inplace=True)

data.head()
```

```python
data['percentage_available'] = data['num_bikes_available'] / (data['num_bikes_available'] + data['num_docks_available'])
output_data = data[['station_id', 'year','month', 'day', 'hour', 'percentage_available']].copy()
output_data.reset_index(drop=False, inplace=True)
output_data.rename(columns={'index': 'index'}, inplace=True)
```

```python
grouped = output_data.groupby(['station_id', 'year', 'month', 'day'])

output_data['ctx-1h'] = grouped['percentage_available'].shift(-12)
output_data['ctx-2h'] = grouped['percentage_available'].shift(-24)
output_data['ctx-3h'] = grouped['percentage_available'].shift(-36)
output_data['ctx-4h'] = grouped['percentage_available'].shift(-48)
```

```python
# Group the data by station_id, month, day, and hour
grouped = data.groupby(['station_id', 'month', 'day', 'hour'])

# Calculate the percentage of available bikes for each station at different time intervals
data['percentage_available'] = data['num_bikes_available'] / (data['num_bikes_available'] + data['num_docks_available'])

# Function to calculate the context for each hour
def calculate_context(group):
    context = {}
    for i in range(1, 5):
        shifted = group['percentage_available'].shift(i)
        context[f'ctx-{i}'] = shifted.fillna(method='bfill')
    return pd.Series(context)

# Apply the function to each group and concatenate the results
contexts = grouped.apply(calculate_context)

# Combine the original DataFrame with the calculated contexts
result = pd.concat([data[['station_id', 'month', 'day', 'hour']], contexts], axis=1)

# Reset index and write to CSV
result.reset_index(drop=True, inplace=True)
result.to_csv('output.csv', index_label='index')
```
